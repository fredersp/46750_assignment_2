```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import os
```

```python
# data path
data_path = os.path.join("..", "data")

# Load ETS data
ets_data = pd.read_csv(os.path.join(data_path, "ETSDailyPrices.csv"), header=1)
```

```python
ets_data.head()
```

```python
# Make a histogram of the primary market prices
plt.hist(ets_data['Primary Market'], bins=30, density=True, alpha=0.6, color='g')
```

```python
# Use historical data to estimate parameters for a normal distribution
mean_price = ets_data['Primary Market'].mean()
std_price = ets_data['Primary Market'].std()

# Generate x values for the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean_price, std_price)

```
